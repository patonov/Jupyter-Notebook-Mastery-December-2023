```python
import pandas as pd

data = pd.DataFrame({'a': np.arange(10),
        'c': np.random.randint(1, 10, 10),
        'd': np.random.randn(10)})

data['b'] = data['a'] + 10 * np.random.randn(10)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('bottom')
plt.ylabel('vertical scale')
plt.show()
```
