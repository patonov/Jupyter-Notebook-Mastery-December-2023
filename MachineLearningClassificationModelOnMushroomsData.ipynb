```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("./input/mushrooms.csv")
```

```python
data["class"].value_counts()
```

```python
target = data["class"]
```

```python
target.value_counts().plot(kind="bar")
```

```python
features = data.drop("class", axis=1)
```

```python
import matplotlib.pyplot as plt
```

```python
for col in features.columns:
    features[col].value_counts().plot(kind="bar")
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
encoder.fit_transform(target)
```

```python
encoded_target = encoder.fit_transform(target)
```

```python
for col in features.columns:
    features[col] = encoder.fit_transform(features[col])
```

```python
features.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
for col in features.columns:
    features[col] = scaler.fit_transform(features[[col]])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, encoded_target, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
accuracy_score(y_test, predictions)
```

```python
conf_matrix = confusion_matrix(y_test, predictions)
```

```python
display = ConfusionMatrixDisplay(conf_matrix)
```

```python
display.plot()
```
